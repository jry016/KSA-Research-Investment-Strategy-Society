```python
df1 = df1.set_index('Date')
df2 = df2.set_index('Date')
df3 = df3.set_index('Date')
df4 = df4.set_index('Date')
df5 = df5.set_index('Date')
df6 = df6.set_index('Date')
df7 = df7.set_index('Date')
df8 = df8.set_index('Date')
df9 = df9.set_index('Date')
```

```python
df1.index=pd.to_datetime(df1.index)
df2.index=pd.to_datetime(df2.index)
df3.index=pd.to_datetime(df3.index)
df4.index=pd.to_datetime(df4.index)
df5.index=pd.to_datetime(df5.index)
df6.index=pd.to_datetime(df6.index)
df7.index=pd.to_datetime(df7.index)
df8.index=pd.to_datetime(df8.index)
df9.index=pd.to_datetime(df9.index)
```

```python
df9
```

```python
df1["Close+1"]=df1["S&P Composite 1500"].shift(1)
df1["Close-1"]=df1["S&P Composite 1500"].shift(-1)
df1[["S&P Composite 1500","Close+1","Close-1"]]
```

```python
df1["dailyReturn"]=(df1["S&P Composite 1500"]-df1["Close+1"])/df1["Close+1"]
```

```python
df1[["S&P Composite 1500","dailyReturn"]]
```

```python
df1["RollMean20"]=df1["dailyReturn"].rolling(20).mean()
df1[["S&P Composite 1500","dailyReturn","RollMean20"]].head(30)
```

```python
df1["RollSum20"]=df1["dailyReturn"].rolling(20).sum()
df1[["S&P Composite 1500","dailyReturn","RollMean20","RollSum20"]].head(30)
```

```python
from pandas.tseries.offsets import MonthEnd
```

```python
df1_monthly = df1.loc[df1.index == df1.index + pd.offsets.MonthBegin(0)]
df1_monthly.to_csv('df1_monthly.csv')
df1_monthly
```

```python
filtered_df1 = df1.loc[(df1.index.day == 6) | (df1.index.day == 8)]
filtered_df1["Month"]=filtered_df1.index.month
filtered_df1["Year"]=filtered_df1.index.year
filtered_df1[["S&P Composite 1500","dailyReturn","RollMean20","RollSum20","Month","Year"]].head(30)
```

```python
filtered_df1 = filtered_df1.drop_duplicates(subset=["Month","Year"])
filtered_df1.to_csv('filtered_df1.csv')
filtered_df1
```
