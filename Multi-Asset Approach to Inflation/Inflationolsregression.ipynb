```python
from statsmodels.regression.rolling import RollingOLS
from statsmodels.datasets import longley
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime, timedelta

%matplotlib inline
plt.style.use('seaborn')

exog = sm.add_constant(final_result['MoM%'])
endo = final_result['S&P Composite 1500']
model = RollingOLS(endo, exog, window = 12)
rolling_res = model.fit()
df1params = rolling_res.params
df1params.head(50)
```

```python
rolling_res.plot_recursive_coefficient(variables=['MoM%'])                     
```

```python
endo_vars = ['S&P Composite 1500','S&P GSCI Copper','S&P GSCI Crude Oil',
             'S&P GSCI Gold','S&P GSCI','S&P U.S. Aggregate Bond Index',
             'S&P U.S. TIPS Index (USD)','S&P United States REIT (USD)' ]

params = pd.DataFrame()

for item in endo_vars:
    exog = sm.add_constant(final_result['MoM%'])
    endo = final_result[item]
    model = RollingOLS(endo, exog, window = 12)
    rolling_res = model.fit()
    rolling_res.params.rename(columns={'MoM%': item},inplace=True)
    params = pd.concat([params,rolling_res.params], axis=1)
    fig = rolling_res.plot_recursive_coefficient(variables='MoM%', figsize=(15,5))
    plt.xlabel('Time Step')
    plt.ylabel('Inflation Beta for ' + item)
    plt.show
    
```

```python
params = params.dropna()
params
```

```python
params.drop(['const'], axis = 1, inplace=True)
params
```

```python
column_names = list(params.columns.values)
column_names
```

```python
# params['S&P Composite 1500'] = [(1 + t) if t >= 0
#          else 1/(1-t) for t in params['S&P Composite 1500']]
# params['S&P Composite 1500'].dtype

for item in endo_vars:
    params[item] = [(1 + t) if t >= 0 else 1/(1-t) for t in params[item]]
    
params
```

```python
params["Aggregated Beta"] = params.sum(axis=1)
params
```

```python
# params["S&P Composite 1500"] = params["S&P Composite 1500"].div(params["Aggregated Beta"].values)
for item in endo_vars:
    params[item] = params[item].div(params["Aggregated Beta"].values)
    
params
```
