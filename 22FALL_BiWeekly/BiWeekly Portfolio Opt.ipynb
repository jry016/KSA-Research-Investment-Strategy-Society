```python
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
```

```python
# Expected Annualized Returns and Annualized Sample Covariance Matrix
mu = expected_returns.mean_historical_return(final_df)
S = risk_models.sample_cov(final_df)
```

```python
# Optimize for Max Sharpe Ratio
ef = EfficientFrontier(mu, S) # Create EfficientFrontier Object
weights = ef.max_sharpe()
cleaned_weights = ef.clean_weights()
print(cleaned_weights)
```

```python
ef.portfolio_performance(verbose = True)
```

```python
from pypfopt import CLA, plotting

cla = CLA(mu, S)
cla.max_sharpe()
cla.portfolio_performance(verbose=True);

ax = plotting.plot_efficient_frontier(cla, showfig=True)
```

```python
# Get Discrete # of Allocation Per Stock
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices
```

```python
port_max_val = 10000
latest_prices = get_latest_prices(final_df)
weights = cleaned_weights
da = DiscreteAllocation(weights, latest_prices, total_portfolio_value = port_max_val)
allocation, leftover = da.greedy_portfolio()
```

```python
print("Portfolio Allocation: ", allocation)
print("Remaining Value: ", leftover)
```
